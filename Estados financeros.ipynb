```python
import yahoofinancials
import pandas as pd
import matplotlib.pyplot as plt
from yahoofinancials import YahooFinancials
```

```python
meta_financials = YahooFinancials('META')
meta_financials
```

```python
balances_trimestrales = meta_financials.get_financial_stmts('quarterly', 'balance')
balances_trimestrales
```

```python
balance_trimestral = balances_trimestrales['balanceSheetHistoryQuarterly']['META'][3]

balance_trimestral_df = pd.DataFrame(balance_trimestral).T

balance_trimestral_df = balance_trimestral_df.applymap(lambda x: '{:,.2f}'.format(x))

balance_trimestral_df
```

```python
balance_trimestral_df.dtypes
```

```python
lista_balance_trimestral = []

for i in range(0,5):
    balance_trimestral = balances_trimestrales['balanceSheetHistoryQuarterly']['META'][i]
    balance_trimestral_df = pd.DataFrame(balance_trimestral).T
    lista_balance_trimestral.append(balance_trimestral_df)
    
balance_trimestral_total = pd.concat(lista_balance_trimestral, axis=0)
balance_trimestral_total
```

```python
balance_trimestral_total.columns.tolist()
```

```python
# Ratio de liquidez corrientes = Activo corriente / pasivo corriente

liquidez = balance_trimestral_total['currentAssets'] /  balance_trimestral_total['currentLiabilities']
liquidez
```

```python
# Ratio de endeudamiento = Deuda total / Capital Total

endeudamiento = balance_trimestral_total['totalDebt'] /  balance_trimestral_total['stockholdersEquity']
endeudamiento
```

```python
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(balance_trimestral_total.index, balance_trimestral_total.workingCapital)

# Establecer el título y las etiquetas de los ejes
ax.set_title('Working Capital por Fecha', fontsize=12)
ax.set_xlabel('Fecha', fontsize=14)
ax.set_ylabel('Working Capital (en millones de u$s)', fontsize=10)

plt.show()
```

```python
# Calcular la variación porcentual entre el primer y el último valor del gráfico
var_pct = (balance_trimestral_total['workingCapital'].iloc[-1] - balance_trimestral_total['workingCapital'].iloc[0]) / balance_trimestral_total['workingCapital'].iloc[0] * 100

# Crear una figura con dos subplots: uno para el gráfico de línea y otro para el gráfico de área
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8))

# Graficar la curva de 'workingCapital' en el primer subplot
ax1.plot(balance_trimestral_total.index, balance_trimestral_total['workingCapital'], marker='o')
ax1.set_title('Working Capital')
ax1.set_ylabel('Working Capital (en millones de u$s)', fontsize=9)

# Rellenar el área debajo de la curva con un color gris claro
ax1.fill_between(balance_trimestral_total.index, balance_trimestral_total['workingCapital'], color='lightgray',
                 alpha=0.4)

# Agregar una flecha desde el primer valor hasta el último valor del gráfico en el primer subplot
ax1.annotate("", xy=(balance_trimestral_total.index[-1], balance_trimestral_total['workingCapital'].iloc[-1] + 0.5e10),
             xytext=(balance_trimestral_total.index[0], balance_trimestral_total['workingCapital'].iloc[0]+ 0.5e10),
             arrowprops=dict(arrowstyle="->", lw=1, color='red'), annotation_clip=False)

start = (balance_trimestral_total.index[0], balance_trimestral_total['workingCapital'].iloc[0] + 0.5e10)

# Agregar un texto encima de la flecha con la variación porcentual en color rojo o verde y con una flecha hacia arriba o hacia abajo según sea el caso
if var_pct < 0:
    ax1.annotate("{:.1f}%\u2193".format(var_pct), xy=(i/2, start[1]), xytext=(i/2, start[1]*0.92), ha="center",
                 color='red')
else:
    ax1.annotate("{:.1f}%\u2191".format(var_pct), xy=(i/2, start[1]), xytext=(i/2, start[1]*0.92), ha="center",
                 color='green')

balance_trimestral_total['workingCapital_pct_change'] = balance_trimestral_total['workingCapital'].pct_change() * 100
    
    
# Graficar la variación porcentual en el segundo subplot con barras de colores
bars = balance_trimestral_total['workingCapital_pct_change']
colors = ['g' if bar >= 0 else 'r' for bar in bars]
ax2.bar(balance_trimestral_total.index, bars, color=colors)
ax2.set_title('Variación porcentual del Working Capital')
ax2.set_xlabel('Fecha')
ax2.set_ylabel('% Cambio', fontsize=9)

# Agregar etiquetas de texto con los valores de la variación porcentual
for i, v in enumerate(bars):
    if v >= 0:
        ax2.annotate("{:.1f}%".format(v), xy=(i, v), xytext=(i, v-1.5), color='white', ha='center')
    else:
        ax2.annotate("{:.1f}%".format(v), xy=(i, v), xytext=(i, v+0.5), color='white', ha='center')

ax1.set_ylim([2e10, 4.5e10])
 
# Ajustar la separación vertical entre los subplots
plt.subplots_adjust(hspace=0.25)  

plt.show()
```
