```python
albums = sp.artist_albums(artist_id, album_type = 'album',limit=50,)


album = []
lanzamiento = []
tracks = []
portada = []


for a in albums['items']:
    album.append(a['name'])
    lanzamiento.append(a['release_date'])
    tracks.append(a['total_tracks'])
    portada.append(a['images'][0]['url'])

    
len(album),len(lanzamiento),len(tracks),len(portada) 
```
