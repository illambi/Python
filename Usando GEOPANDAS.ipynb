```python
import geopandas
```

```python
#leer el dataset de Nueva York
path_to_data = geopandas.datasets.get_path("nybb")
gdf = geopandas.read_file(path_to_data)
print(gdf.shape)
gdf
```

```python
#cambiar el Ã­ndice
gdf = gdf.set_index("BoroName")
gdf.head()
```

```python
#medir el Ã¡rea
gdf["area"] = gdf.area
gdf.head()
```

```python
gdf['centroid'] = gdf.centroid
gdf.head()
```

```python
#distancia a Brooklyn con el centroide
brook_point = gdf['centroid'].iloc[2]
gdf['distance'] = gdf['centroid'].distance(brook_point)
gdf['distance']
```

```python
gdf.plot('area', cmap='Blues')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()

gdf.plot(ax=ax, color='white', edgecolor='black')
gdf['centroid'].plot(ax=ax)
plt.show()
```

```python
import pandas as pd

florence = pd.read_csv('florence.csv')
florence.head()
```

```python
florence = florence.drop(['AdvisoryNumber', 'Forecaster', 'Received'], axis=1)
florence.head()
```

```python
florence['Long'] = 0 - florence['Long']
florence.head()
```

```python
florence['coordinates'] = florence[['Long', 'Lat']].values.tolist()
florence.head()
```

```python
from shapely.geometry import Point

florence['coordinates'] = florence['coordinates'].apply(Point)
florence.head()
```

```python
florence = geopandas.GeoDataFrame(florence, geometry='coordinates')
florence.head()
```

```python
type(florence)
```

```python
florence.Wind.mean()
```

```python
florence.Wind.max()
```
