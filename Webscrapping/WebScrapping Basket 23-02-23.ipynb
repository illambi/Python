```python
url = 'https://www.basketball-reference.com/leagues/NBA_2019_per_game.html'
```

```python
df = pd.read_html(url, header=0)
df
```

```python
df2019 = df[0]
df2019.head()
```

# 1 Desafio hacer webscrapping a otro año.

# 2 Desafio (opcional) hacer un analisis del mismo

```python
df2019.shape
```

```python
df2019['Year'] = 2019
df2019.head()
```

```python
url2 = 'https://www.basketball-reference.com/leagues/NBA_2020_per_game.html'
df2 = pd.read_html(url2, header = 0)
df2020 = df2[0]

df2020['Year'] = 2020
df2020.head()
```

```python
# Uno los DataFrame (uno debajo del otro)

df1920 = pd.concat([df2019, df2020])
df1920
```

```python
df1920.isnull().sum()
```

```python
#df1920.Player.unique
df1920['Player'].nunique()
```

```python
pos = df1920.groupby(['Pos'])['Player'].nunique().reset_index().sort_values(by='Player', ascending = False)
pos
```

```python
pos2 = df1920.groupby(['Pos'])['PTS'].mean().reset_index().sort_values(by='PTS', ascending = False)
pos2
```

```python
df1920['PTS'] = df1920.PTS.apply(pd.to_numeric, errors='coerce')
```

```python
final = pd.merge(pos, pos2, on='Pos', how='inner')

final.rename({'Pos':'Posicion', 'Player':'Cantidad de Jugadores', 'PTS':'Promedio de Puntos'}, axis=1, inplace=True)


final
```
