```python
url = requests.get ('https://listado.mercadolibre.com.ar/monitores')
url.status_code
```

```python
soup = BeautifulSoup(url.content, 'html.parser')
```

```python
titulos= soup.find_all('h2', attrs={"class":'ui-search-item__title ui-search-item__group__element shops__items-group-details shops__item-title'})
```

```python
titulos[0].text
```

```python
len(titulos)
```

```python
for i in titulos:
  print(i.text)
```

```python
titulos = [i.text for i in titulos]
titulos
```

```python
urls = soup.find_all('a', attrs={'class':'ui-search-result__content ui-search-link'})
```

```python
urls[0].get('href')
```

```python
urls = [i.get('href') for i in urls]
```

```python
print(len(titulos))
len(urls)
```

```python
df = pd.DataFrame({'titulos':titulos,'url':urls})
df['fecha_obs'] = date.today()
df
```

```python
#buscando con la consola encontr la clase div que muestra el precio final pero le suma otras cosas que no me servan

precios = soup.find_all('div', attrs={'class':'ui-search-price__second-line shops__price-second-line'})
```

```python
#son 48 precios como la cantidad de productos 

len(precios)
```

```python
#Esto es lo que encuentra, el precio final pero le agrega otras cosas que no me sirven

precios[0].text
```

```python
#guardo los resultados que me trajo en una lista

precios = [i.text for i in precios]
precios
```

```python
#y con un For recorro esa lista trayendo la primer palabra antes del espacio y la guardo en una nueva lista en donde solo van 
#a quedar los precios. Tambin le agregu el signo $ en cada recorrido que hace el For. Y listo, ya tengo los precios de cada
#producto

precios_separados = []

for t in precios:
    precios_separados2 = t.split()[0]
    precios_separados.append("$ " + precios_separados2)

print(precios_separados)
```

```python
dfpre = pd.DataFrame({'titulos':titulos,'url':urls,'precios':precios_separados})
dfpre['fecha'] = date.today()
dfpre
```
